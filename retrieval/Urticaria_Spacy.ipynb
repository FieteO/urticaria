```python
import json
import spacy
import random
from spacy.scorer import Scorer
from spacy.training.example import Example
from spacy import displacy
```

```python
#Data preperation for Spacy Modeltraining

labeled_data = []
with open(r"./Data/admin.jsonl", "r") as read_file:
    lines = read_file.readlines()
    
    for line in lines:
        line =json.loads(line)
        if "label" in line:
            line["entities"] = line.pop("label")
        else:
            line["entities"] = []

        tmp_ents = []
        for e in line["entities"]:
            if e[2] in ['TREATMENT', 'BIOMARKER', 'CONDITION']:
                #tmp_ents.append({"start": e[0], "end": e[1], "label": e[2]})
                tmp_ents.append((e[0], e[1], e[2]))
        
            line["entities"] = tmp_ents

        if (len(line["data"]) > 5):
            #print (json.dumps({"text": line["data"],"entities": line["entities"]}))
            #labeled_data.append(json.dumps({"text": line["data"],"entities": line["entities"]}))
            labeled_data.append({"text": line["data"],"entities": line["entities"]})

DATA = []
for entry in labeled_data:
    entities = []
    for e in entry['entities']:
        entities.append((e[0], e[1],e[2]))
    spacy_entry = (entry['text'], {"entities": entities})
    DATA.append(spacy_entry)
```

```python
# Train-Test-Splitting
TRAINING_DATA = DATA[-int(len(DATA)*0.75):]
TEST_DATA = DATA[:-int(len(DATA)*0.75)]
```

```python

nlp = spacy.blank("en")
ner = nlp.create_pipe("ner")
nlp.add_pipe('ner')
ner.add_label("BIOMARKER")
ner.add_label("TREATMENT")
ner.add_label("CONDITION")

# Start the training
nlp.begin_training()

# Loop for 40 iterations
for itn in range(40):
    # Shuffle the training data
    random.shuffle(TRAINING_DATA)
    losses = {}
# Batch the examples and iterate over them
    for batch in spacy.util.minibatch(TRAINING_DATA, size=2):
        #texts = [text for text, entities in batch]
        #annotations = [entities for text, entities in batch]
        for text,annotations in batch:
            doc = nlp.make_doc(text)
            example = Example.from_dict(doc, annotations)
        
# Update the model
        nlp.update([example], losses=losses, drop=0.3)
        print(f'losses: {losses}')
        
```

```python
#Save the Model
nlp.to_disk("./Model/urticaria.model")
```

```python
def evaluate(ner_model, testdata):
    scorer = Scorer(ner_model)
    examples = []
    for input_, annot in testdata:
        
        doc = ner_model(input_)
        example = Example.from_dict(doc, annot)
        example.predicted = ner_model(str(example.predicted))
        examples.append(example)
    
    results = scorer.score(examples)
    return results

ner_model = spacy.load("./Model/urticaria.model")
results = evaluate(ner_model, TEST_DATA)
```
